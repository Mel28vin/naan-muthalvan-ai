### Outlier Replacement (Median)

```python
q1 = df["living_area_renov"].quantile(0.25)
q3 = df["living_area_renov"].quantile(0.75)
```

```python
print(q1, q3)
```

```python
IQR=q3-q1
```

```python
Upper_limit = q3+1.5 * IQR
lower_limit = q1-1.5 * IQR
```

```python
print(Upper_limit)
```

```python
df.median()
```

```python
df["living_area_renov"]=np.where(df["living_area_renov"]>Upper_limit, df.median()["living_area_renov"],df["living_area_renov"])
```

```python
sns.boxplot(df["living_area_renov"])
```
