### Outlier Replacement (Median)

```python
q1 = df["Area of the house(excluding basement)"].quantile(0.25)
q3 = df["Area of the house(excluding basement)"].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
Upper_limit = q3+1.5 * IQR
lower_limit = q1-1.5 * IQR
```

```python
Upper_limit
```

```python
df['Area of the house(excluding basement)']=np.where(df['Area of the house(excluding basement)']>Upper_limit,df.median()["Area of the house(excluding basement)"],df['Area of the house(excluding basement)'])
```

```python
sns.boxplot(df['Area of the house(excluding basement)'])
```
